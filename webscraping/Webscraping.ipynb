# Get places in western cape


```python
# IF YOU CAN't FIND TABLE
# for table in soup.find_all('table'):
#   print(table.get('class'))

df_list = []
```

```python

u = "https://en.wikipedia.org/wiki/Hanover_Park,_Cape_Town"
response = requests.get(u)
soup = BeautifulSoup(response.text, "html.parser")


import pandas as pd 

table = soup.find('table', attrs={'class':"nowraplinks"})
table_body = table.find('tbody')

rows = table_body.find_all('tr')

temp =[]
for idx,row in enumerate(rows):
#     print idx, row.text

    dict = {}
    dict['Community']  = row.text.split('\n')[0]
    dict['Suburb']  = row.text.split('\n')
        
    dict['Suburb'] = dict['Suburb'][1:] 
    temp.append(dict)
        
        
import numpy as np
df = pd.DataFrame(temp)
lst_col = 'Suburb'

df1 = pd.DataFrame({
      col:np.repeat(df[col].values, df[lst_col].str.len())
      for col in df.columns.drop(lst_col)}
    ).assign(**{lst_col:np.concatenate(df[lst_col].values)})[df.columns]

df1.replace('', np.nan, inplace=True)
df1.dropna(subset=['Suburb'], inplace=True)

df1["Municipality"] = 'Cape Town' 
df1["Province"] = 'Western Cape' 


df_list.append(df1)


df1.loc[df1['Suburb'] == 'Hanover Park']
```

```python
u = "https://en.wikipedia.org/wiki/City_of_Johannesburg_Metropolitan_Municipality"
response = requests.get(u)
soup = BeautifulSoup(response.text, "html.parser")


import pandas as pd 

# for table in soup.find_all('table'):
#   print(table.get('class'))

# FOR REGIONS
table = soup.find_all('table', attrs={'class':"nowraplinks mw-collapsible mw-collapsed navbox-subgroup"})
 
# for i in table:
#     print i.text

 
temp =[]
for idx,row in enumerate(table):
#     print idx, row.text

    dict = {}
    dict['Community']  = row.text.split('\n')[0]
    dict['Suburb']  = row.text.split('\n')
        
    dict['Suburb'] = dict['Suburb'][1:] 
    temp.append(dict)
        
# print temp
        
import numpy as np
df = pd.DataFrame(temp)
lst_col = 'Suburb'

df1 = pd.DataFrame({
      col:np.repeat(df[col].values, df[lst_col].str.len())
      for col in df.columns.drop(lst_col)}
    ).assign(**{lst_col:np.concatenate(df[lst_col].values)})[df.columns]

df1.replace('', np.nan, inplace=True)
df1.dropna(subset=['Suburb'], inplace=True)

df1["Municipality"] = 'City of Johannesburg' 
df1["Province"] = 'Gauteng' 

df_list.append(df1)

 
```
