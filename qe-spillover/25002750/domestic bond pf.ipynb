```python
import scipy 
import numpy as np

import pandas as pd

max = pd.read_csv('df_final_max25002750.csv')
max['cal'] = 'max'
max= max[max['QE']!='x_QE0']

print np.unique(max.QE)

max.QE = max.QE.str.replace('x_QE0','0')
max.QE =max.QE.str.replace('x_QE100','100')
max.QE =max.QE.str.replace('x_QE200','200')
max.QE =max.QE.str.replace('x_QE400','400')
max.QE =max.QE.str.replace('x_QE600','600')
max.QE =max.QE.str.replace('x_QE800','800')

max.QE =max.QE.str.replace('x_QE1000','1000')
max.QE =max.QE.str.replace('x_QE1200','1200')
max.QE =max.QE.str.replace('x_QE1400','1400')
max.QE =max.QE.str.replace('x_QE1600','1600')
max.QE =max.QE.str.replace('x_QE1800','1800')
max.QE =max.QE.str.replace('x_QE2000','2000')
max.QE =max.QE.str.replace('x_QE2200','2200')

max.QE =max.QE.str.replace('x_QE2400','2400')
max.QE =max.QE.str.replace('x_QE2600','2600')
 
 
max['QE'] = pd.to_numeric(max['QE'])
max = max[max.QE!=0]


min = pd.read_csv('df_final_min25002750.csv')
min['cal'] = 'min'

print np.unique(min.QE)
min.QE =min.QE.str.replace('n_QE','')
min['QE'] = pd.to_numeric(min['QE'])
min = min[min.QE!=0]

med = pd.read_csv('df_final_med25002750.csv')
med['cal'] = 'med'
print np.unique(med.QE)
med.QE =med.QE.str.replace('d_QE','')
med['QE'] = pd.to_numeric(med['QE'])
med = med[med.QE!=0]


med['dom_bond_yields_perQE'] = med['funds[0].exp.returns[portfolios[0]]']*100*250*100/med.QE
med['dom_bond_yields'] = med['funds[0].exp.returns[portfolios[0]]']*100*250*100 
ymed1 =med['dom_bond_yields'].as_matrix()
ymed2 =med['dom_bond_yields_perQE'].as_matrix()

max['dom_bond_yields_perQE'] = max['funds[0].exp.returns[portfolios[0]]']*100*250*100/max.QE
max['dom_bond_yields'] = max['funds[0].exp.returns[portfolios[0]]']*100*250*100 
ymax1 =max['dom_bond_yields'].as_matrix()
ymax2 =max['dom_bond_yields_perQE'].as_matrix()

min['dom_bond_yields'] = min['funds[0].exp.returns[portfolios[0]]']*100*250*100
min['dom_bond_yields_perQE'] = min['funds[0].exp.returns[portfolios[0]]']*100*250*100/min.QE

ymin1 =min['dom_bond_yields'].as_matrix()
ymin2 =min['dom_bond_yields_perQE'].as_matrix()



import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9,3))

# generate some random test data
all_data = [ymin1,ymed1,ymax1]
all_dataperQE = [ymin2,ymed2,ymax2]


# plot violin plot
axes[0].violinplot(all_data,
                   showmeans=False,
                   showmedians=True)
axes[0].set_title('Effect for all observations \n t>2500 ')
axes[0].set_ylabel('basis points')
 

# adding horizontal grid lines
for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(len(all_data))])
    ax.set_xlabel('Domestic bond yield effect')

# plot box plot


violin_parts =axes[1].violinplot(all_dataperQE,
                   showmeans=False,
                   showmedians=True)

for pc in violin_parts['bodies']:
    pc.set_facecolor('#D43F3A')
    pc.set_edgecolor('black')
    pc.set_alpha(1)


axes[1].set_title('Effect per QE \n t>2500')
    
# add x-tick labels
plt.setp(axes, xticks=[y+1 for y in range(len(all_data))],
         xticklabels=['Min', 'Med', 'Max'])
plt.show()

print 'Cal  ', '5%', '   Mean', '   95%'
print 'Min', round(np.percentile(all_dataperQE[0].tolist(),5),3), round(all_dataperQE[0].mean(),3), round(np.percentile(all_dataperQE[0].tolist(),95),3)
print 'Med' , round(np.percentile(all_dataperQE[1].tolist(),5),3), round(all_dataperQE[1].mean(),3), round(np.percentile(all_dataperQE[1].tolist(),95),3)
print 'Max', round(np.percentile(all_dataperQE[2].tolist(),5),3), round(all_dataperQE[2].mean(),3), round(np.percentile(all_dataperQE[2].tolist(),95),3)

# print np.perncentile([all_dataperQE[2], 95) 
```
