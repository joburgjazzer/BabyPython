```python
def freq1way(df, var):
    """
    Returns 1-way frequency table (counts and percentages) for column var of pandas dataframe df
    """
    f = df[var].value_counts().sort_index().to_frame(var)
    f['pct'] = 100 * f[var] / f[var].sum()
    return f
```

```python
# Example call of function freq1way
fQE = freq1way(df, 'QE')
fQE
```
